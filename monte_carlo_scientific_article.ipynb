## The Monte Carlo Method:

By definition:

> The Monte Carlo method is a computational algorithm that uses repeated random sampling to obtain numerical results, particularly for problems involving probability and uncertainty.It approximates solutions to complex integrals, optimizations, and predictions by simulating many possible outcomes based on probability distributions.

### What Is a Monte Carlo Simulation?
**A Monte Carlo simulation** is a way to model the probability of different outcomes in a process that cannot easily be predicted due to the intervention of **random variables**. It is a technique used to understand the impact of risk and uncertainty.

## Hisrory:

> __A game of solitaire, a case of insomnia, a chance meeting at a train station, and a penchant for gambling all contributed to the creation of one of the most influential computing tools in the world: the Monte Carlo method.__ 

#### Where does it get its name from?
The Monte Carlo simulation was named after the famous gambling destination in Monaco because chance and random outcomes are central to this modeling technique, as they are to games like roulette, dice, and slot machines.

#### Where and by who was the method invented?
The technique was initially developed by [Stanislaw Ulam](https://en.wikipedia.org/wiki/Stanis%C5%82aw_Ulam), a mathematician who worked on the [Manhattan Project](https://en.wikipedia.org/wiki/Manhattan_Project), the secret effort to create the first atomic weapon.He shared his idea with [John Von Neumann](https://en.wikipedia.org/wiki/John_von_Neumann), [Robert Richtmyer](https://en.wikipedia.org/wiki/Robert_D._Richtmyer) and [Nick Metropolis](https://en.wikipedia.org/wiki/Nicholas_Metropolis), colleagues at the Manhattan Project, and collaborated to refine the Monte Carlo simulation.

As Ulam recounts his inspiration:
> The first attempts I made to practice the Monte Carlo Method came from a question that occurred to me in 1946 while I was recovering from an illness and playing solitaire: what are the chances that a Canfield solitaire with 52 cards will come out successfully? After trying to estimate them through combinatorial calculations, I wondered whether a more practical method than “abstract thinking” might be to lay it out a hundred times and simply count the successful plays. \
With the emergence of **fast computers**, this approach seemed possible.Later in 1946, I described the idea to John von Neumann, and we began planning actual calculations.

#### What's the idea behind it?
The **core idea** behind the creation of the Monte Carlo method was to solve mathematically complex or intractable problems, particularly those involving **randomness or high-dimensional variables**.Invented during World War II, the goal was to harness the power of emerging early computers to transform deterministic problems into probabilistic ones (which can be estimated by **observing thousands of random outcomes**).


#### How does it work?
Unlike a **normal forecasting model**, Monte Carlo Simulation predicts a set of outcomes based on an estimated range of values versus a set of fixed input values. In other words, a Monte Carlo Simulation builds a model of possible results by leveraging a **probability distribution**, such as a **uniform or normal distribution**, for any variable that has inherent uncertainty. It, then, recalculates the results over and over, each time using a different set of random numbers between the minimum and maximum values. In a typical Monte Carlo experiment, this exercise can be repeated **thousands of times** to produce a large number of **likely outcomes**.

##### Basic Monte Carlo Simulation Algorithm:
The following pseudo-code algorithm demonstrates the general workflow of a Monte Carlo simulation. The goal is to estimate a quantity (such as an expected value, probability, or integral) by repeated random sampling.
```
    Input:
        N → number of simulations(A very big number)

    total ← 0

    for i = 1 to N do
        x ← generate random variable from probability distribution
        y ← evaluate function f(x)
        total ← total + y
    end for

    estimate ← total / N
    return estimate
```

Monte Carlo simulation approximates a value using:
1. **Random Sampling**
2. **Repeated Experiments**
3. **Averaging**
4. **Approximate Solution**


#### How to use Monte Carlo method?

Regardless of what tool you use or for what purpose you are using it, \
The Monte Carlo method involves **three basic steps**:

1. **Set up the predictive model** \
Identify both the dependent variable to be predicted and the independent variables (also known as the input, risk or predictor variables) that will drive the prediction.

2. **Specify probability distributions of the independent variables.** \
Use historical data and the analyst’s subjective judgment to define a range of likely values and assign probability weights for each.

3. **Run simulations repeatedly, generating random values of the independent variables.** \
Do this until enough results are gathered to make up a representative sample of the near infinite number of possible combinations.


After we know the workflow of the Monte Carlo method, lets try to give some example and represent that example with a basic algorithm which estimates specific results:

One simple example of a Monte Carlo Simulation is to consider calculating the probability of rolling two standard dice. There are 36 combinations of dice rolls. Based on this, we can manually compute the probability of a particular outcome.

Using a Monte Carlo Simulation, you can simulate rolling the dice 10,000 times (or more) to achieve more accurate predictions.

````python
import numpy as np
````

By following the steps we have specified earlier, we will make an algorithm, which calculates the probability that the sum of the two dices would be 7. \
The theoretical probability of rolling 7 with two dice is:
$$ 6 / 36 \approx 0.1667 $$
But lets see it with code:

````python
import numpy as np

# Number of simulations - should be a very big number in order the results to be more accurate!
n = 10000

# Counter for successful outcomes
success_count = 0

# Target event
target_sum = 7

for i in range(n):

    # Generate random values for independent variables
    die1 = np.random.randint(1,7)
    die2 = np.random.randint(1,7)

    # Evaluate the predictive model
    dice_sum = die1 + die2

    # Check if the event occurs
    if dice_sum == target_sum:
        success_count += 1

# Estimate probability
estimated_probability = success_count / n

print("Estimated probability:", estimated_probability)
````

#### Lets explain the algorithm a little bit more detailed:
##### 1.Number of Simulations (N):
` N = 10000 `

Larger values of N produce:

- more stable estimates
- smaller statistical error

##### 2.Random Sampling:
This function simulates rolling a die: `np.random.randint(1,7)`

- Each call generates an independent random value.

##### 3.Model Evaluation:
This step represents the system being studied: `dice_sum = die1 + die2`

##### 4.Event Detection:
This checks whether the desired outcome occurred: `if dice_sum == target_sum`

##### 5.Counting Successful Outcomes:
Each time the event occurs, the counter increases: `success_count += 1`

##### 6.Probability Estimation:
This computes the relative frequency of the event: `estimated_probability = success_count / N`


Now let's create the same algorithm but a little bit more compacted and using the build-in numpy functions for estimating average values:

````python
n = 10000

first_dice = np.random.randint(1, 7, n)
second_dice = np.random.randint(1, 7, n)

dice_sum = first_dice + second_dice

prob_of_sum_being_two = np.mean(dice_sum == 2)  # Calculates the arithmetic average of the specified elements.
print(prob_of_sum_being_two)
````

The `mean()` function is typically used to compute the average of an entire array or along a specific axis, helping to summarize large datasets with a single representative number.

Let's test what will be the probability of the sum being even:

````python
prob_of_sum_being_even = np.mean(dice_sum % 2 == 0)
print(prob_of_sum_being_even)
````

Or maybe being more then or equal to 10:

````python
prob_of_sum_being_more_or_equal_to_ten = np.mean(dice_sum >= 10)
print(prob_of_sum_being_more_or_equal_to_ten)
````

# The Mathematical foundations:

**The Monte Carlo Method** is built on several important ideas from probability theory and statistics.Although the method itself is simple in principle, using random sampling and repeated experiments to approximate a result, its reliability and accuracy depend on these underlying mathematical concepts.Understanding them, helps explain why Monte Carlo simulations work, how their results should be interpreted, and what factors influence their precision and accuracy.

In this section, we will introduce the **key mathematical ideas** which form the foundation of the Monte Carlo methods.

1. **Random Variables** \
    – variables whose values depend on random outcomes.
2. **Probability Distributions** \
    – the patterns which form the shape of randomeness.
3. **Expected Value** \
    – the average value of a random variable over many trials.
4. **Variance and Standard Deviation** \
    – measure how much simulation results vary.
5. **The Law of Large Numbers** \
    – the principle that the average of many random samples approaches the true expected value.
6. **The Central Limit Theorem** \
    – a result that explains how the distribution of sample averages behaves when many simulations are performed.
7. **Random Sampling** \
    – the process of generating values from probability distributions to represent uncertainty.
8. **Statistical Estimation** \
    – using simulation results to approximate unknown probabilities or quantities.
