## Hisrory:

> __A game of solitaire, a case of insomnia, a chance meeting at a train station, and a penchant for gambling all contributed to the creation of one of the most influential computing tools in the world: the Monte Carlo method.__ 

#### Where does it get its name from?
The Monte Carlo simulation was named after the famous gambling destination in Monaco because chance and random outcomes are central to this modeling technique, as they are to games like roulette, dice, and slot machines.

#### Where and by who was the method invented?
The technique was initially developed by [Stanislaw Ulam](https://en.wikipedia.org/wiki/Stanis%C5%82aw_Ulam), a mathematician who worked on the [Manhattan Project](https://en.wikipedia.org/wiki/Manhattan_Project), the secret effort to create the first atomic weapon.He shared his idea with [John Von Neumann](https://en.wikipedia.org/wiki/John_von_Neumann), [Robert Richtmyer](https://en.wikipedia.org/wiki/Robert_D._Richtmyer) and [Nick Metropolis](https://en.wikipedia.org/wiki/Nicholas_Metropolis), colleagues at the Manhattan Project, and collaborated to refine the Monte Carlo simulation.

As Ulam recounts his inspiration:
> The first attempts I made to practice the Monte Carlo Method came from a question that occurred to me in 1946 while I was recovering from an illness and playing solitaire: what are the chances that a Canfield solitaire with 52 cards will come out successfully? After trying to estimate them through combinatorial calculations, I wondered whether a more practical method than “abstract thinking” might be to lay it out a hundred times and simply count the successful plays. \
With the emergence of **fast computers**, this approach seemed possible.Later in 1946, I described the idea to John von Neumann, and we began planning actual calculations.

#### What's the idea behind it?
The **core idea** behind the creation of the Monte Carlo method was to solve mathematically complex or intractable problems, particularly those involving **randomness or high-dimensional variables**.Invented during World War II, the goal was to harness the power of emerging early computers to transform deterministic problems into probabilistic ones (which can be estimated by **observing thousands of random outcomes**).


#### How does it work?
Unlike a **normal forecasting model**, Monte Carlo Simulation predicts a set of outcomes based on an estimated range of values versus a set of fixed input values. In other words, a Monte Carlo Simulation builds a model of possible results by leveraging a **probability distribution**, such as a **uniform or normal distribution**, for any variable that has inherent uncertainty. It, then, recalculates the results over and over, each time using a different set of random numbers between the minimum and maximum values. In a typical Monte Carlo experiment, this exercise can be repeated **thousands of times** to produce a large number of **likely outcomes**.

##### Basic Monte Carlo Simulation Algorithm:
The following pseudo-code algorithm demonstrates the general workflow of a Monte Carlo simulation. The goal is to estimate a quantity (such as an expected value, probability, or integral) by repeated random sampling.
```
    Input:
        N → number of simulations(A very big number)

    total ← 0

    for i = 1 to N do
        x ← generate random variable from probability distribution
        y ← evaluate function f(x)
        total ← total + y
    end for

    estimate ← total / N
    return estimate
```

Monte Carlo simulation approximates a value using:
1. **Random Sampling**
2. **Repeated Experiments**
3. **Averaging**
4. **Approximate Solution**


#### How to use Monte Carlo method?

Regardless of what tool you use or for what purpose you are using it, \
The Monte Carlo method involves **three basic steps**:

1. **Set up the predictive model** \
Identify both the dependent variable to be predicted and the independent variables (also known as the input, risk or predictor variables) that will drive the prediction.

2. **Specify probability distributions of the independent variables.** \
Use historical data and the analyst’s subjective judgment to define a range of likely values and assign probability weights for each.

3. **Run simulations repeatedly, generating random values of the independent variables.** \
Do this until enough results are gathered to make up a representative sample of the near infinite number of possible combinations.

One simple example of a Monte Carlo Simulation is to consider calculating the probability of rolling two standard dice. There are 36 combinations of dice rolls. Based on this, we can manually compute the probability of a particular outcome.

Using a Monte Carlo Simulation, you can simulate rolling the dice 10,000 times (or more) to achieve more accurate predictions.

````python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
````

By following the steps we have specified earlier, we will make an algorithm, which calculates the probability that the sum of the two dices would be 7. \
The theoretical probability of rolling 7 with two dice is:
$$ 6 / 36 \approx 0.1667 $$
But lets see it with code:

````python
# Number of simulations - should be a very big number in order the results to be more accurate!
n = 10000

# Counter for successful outcomes
success_count = 0

# Target event
target_sum = 7

for i in range(n):

    # Generate random values for independent variables
    die1 = np.random.randint(1,7)
    die2 = np.random.randint(1,7)

    # Evaluate the predictive model
    dice_sum = die1 + die2

    # Check if the event occurs
    if dice_sum == target_sum:
        success_count += 1

# Estimate probability
estimated_probability = success_count / n

print("Estimated probability:", estimated_probability)
````

#### Lets explain the algorithm a little bit more detailed:
##### 1.Number of Simulations (N):
` N = 10000 `

Larger values of N produce:

- more stable estimates
- smaller statistical error

##### 2.Random Sampling:
This function simulates rolling a die: `np.random.randint(1,7)`

- Each call generates an independent random value.

##### 3.Model Evaluation:
This step represents the system being studied: `dice_sum = die1 + die2`

##### 4.Event Detection:
This checks whether the desired outcome occurred: `if dice_sum == target_sum`

##### 5.Counting Successful Outcomes:
Each time the event occurs, the counter increases: `success_count += 1`

##### 6.Probability Estimation:
This computes the relative frequency of the event: `estimated_probability = success_count / N`

# The Mathematical foundations:

**The Monte Carlo Method** is built on several important ideas from probability theory and statistics.Although the method itself is simple in principle, using random sampling and repeated experiments to approximate a result, its reliability and accuracy depend on these underlying mathematical concepts.Understanding them, helps explain why Monte Carlo simulations work, how their results should be interpreted, and what factors influence their precision and accuracy.

In this section, we will introduce the **key mathematical ideas** which form the foundation of the Monte Carlo methods.

1. **Random Variables** \
    – variables whose values depend on random outcomes.
2. **Probability Distributions** \
    – the patterns which form the shape of randomeness.
3. **Expected Value** \
    – the average value of a random variable over many trials.
4. **Variance and Standard Deviation** \
    – measure how much simulation results vary.
5. **The Law of Large Numbers** \
    – the principle that the average of many random samples approaches the true expected value.
6. **The Central Limit Theorem** \
    – a result that explains how the distribution of sample averages behaves when many simulations are performed.
7. **Random Sampling** \
    – the process of generating values from probability distributions to represent uncertainty.
8. **Statistical Estimation** \
    – using simulation results to approximate unknown probabilities or quantities.

To start with, let's get acquainted with one of the many Monte Carlo techniques, which actually is a
foundational type of Monte Carlo method, and it is considered one of the most fundamental and frequently used techniques within the wide range of algorithms.

Take for an example, calculating the **average** height of the adult population of a country.It would require measuring the height of each person making up that population, summing up the numbers, and dividing them by the total number of people measured.This task might take a **very long time**!What we can do instead is take a sample of that population and compute its average height. It is unlikely to give you the exact average height of the entire population, however, this technique gives a result that is generally a good **approximation** of what that number might be.

Funnily enough, the **approximation** and the exact average of the entire population might sometimes be the same. This is only due to **chance**.In most cases, the numbers will be different.One question we might want to ask though, is **how different** - two key factors determine this difference, which we will address a little later?In fact, as the size of the sample **increases**, this **approximation** converges to the **exact** number(there is a specific law explaining this regularity and we will talk about it a little bit later).In other words, the error between the **approximation** and the actual result is getting smaller as the sample size increases!

Note that the height of a person is a random number.It can be anything really which is the nature of random things. Interestingly enough, a sum of random numbers is another random number. If you can't predict what the number making up the sum are, how can you predict the result of their sum? So, the result of the sum is also a random number - which random number points to the height of the population!

For a mathematician, the height of a population would be called a **random variable**, because the height among people making up this population varies randomly.We generally denote random variables with upper case letters. The letter $X$ is commonly used.

In statistics, the elements making up that population, which as suggested before are random, are denoted with a lower capital letter, generally $x$ as well.All these x's can also be seen as the possible outcomes of the random variable X. If we call X this random variable (the population height), we can express the concept of approximating the average adult population height from a sample with the following pseudo-mathematical formula:

$$
\text{Approximation}(\text{Average}(X)) = \frac{1}{N} \sum^{N}_{n = 1} x_n
$$

Which you can read as, the approximation of the **average value** of the random variable $X$, (the height of the adult population of a given country), is equal to the sum (the $\sum$ sign) of the height of $N$ adults randomly chosen from that population (the samples), divided by the number $N$ (the sample size). This, in essence, is what we call a **Monte Carlo approximation** - a method of **approximating a quantity of interest** — such as an **expected value**.

This **expected value** in statistics corresponds to the **average** of the random variable $X$ and it is called an expectation and is written as $E(X)$.

To summarize, Monte Carlo approximation (which is one of the MC methods) is a technique to approximate the expectation of random variables, using samples. It can be defined mathematically with the following formula:

$$
    E(X) \approx \frac{1}{N} \sum^{N}_{n = 1} x_n
$$

where:
- $X$ is the random variable, which points to the expected(average) height of the population($N$),
- $x_n$ are the random sampled heights,
- $N$ is the number of samples

The mathematical sign $\approx$ means that the formula on the right inside of this sign only gives an *approximation* of what the random variable $X$ expectation $E(X)$ is!

So now let's represent our example, with the population and their average height, with actual code and see what assumptions we can gather from it!:

First, as we have did earlier we should set up the predictive model and specify the random variable!

````python
# The average height of a person of the population.
mean_height = 170

# The standard deviation, which is a required to be specified, in order to sample from a normal distribution.
# We will talk about distributions later on this section!
sd_height = 10

# Random variable: H ~ Normal(mean, std)
# This function generates random values from a normal distribution.
def sample_height(n):  # Here n is the size of the sample.
    # We specify probability distributions of the independent variables.
    return np.random.normal(mean_height, sd_height, n)
````

Each generated value of the function is a height of a randomly selected person from the population!

````python
# We now approximate the expected value 𝐸[X] using random samples.
def get_sample_expected_value(N):
    heights = sample_height(N)
    expected_value = np.mean(heights)
    return expected_value
````

Now that we have defined the functions for creating the random samples and getting the expected value from them, we should make some experiments to see some results:

````python
# The size of the samples
n = 1000

estimate = get_sample_expected_value(n)

print("Estimated average height:", estimate)
print("True average height:", mean_height)
````

As we can see, the estimate(the expected value) is very close to the true mean.So here we got a good approximation of the real average height of the population!From this observation we can assump that when we sample **many** different random values(heights), and here the keyword is **many**, from a normal distribution, we can expect that the average(expected value) will be closely enough to the real mean of the sample(the population)!

Now let's make some experiment and see what will happen if, the size of the sample, differs:

````python
# Here we create an array with different sizes, which are going to be used for the creation of samples.
sample_sizes = np.arange(10, 5000, 10)  # The arrange function, accept for parameters: min/max values and a step by which the numbers to be taken!

estimates = []

# Here for each of the different sizes we calculate the estimated mean and append it to an array.
for n in sample_sizes:
    estimate = get_sample_expected_value(n)
    estimates.append(estimate)
````

So in order to see the difference in the expected values, based on the sizes applied for their samples, we will make a simple plot, to visualize the differences:

````python
# Adjusting the plot figure
plt.figure(figsize=(10,6))

# PLotting the estimated values of the samples and the corresponding sizes applied for them:
plt.plot(sample_sizes, estimates, label="Monte Carlo estimate")
# Drawing the line representing the true population mean - 170:
plt.axhline(mean_height, linestyle="--", label="True population mean")

plt.xlabel("Number of Samples")
plt.ylabel("Estimated Average Height")

plt.title("Monte Carlo Approximation of the Expected Height")
plt.legend()

plt.show()
````

This visualization shows how the **Monte Carlo approximation** approaches the true expected value as the number of samples increases.

With this experiment we can assump that:
> As the number of samples increases, the sample mean approaches the expected value.

This statement is strongly connected to something called: *The Law of large Numbers* - a super important mathematical concept, which we are going to explain a little bit later - this law is crucial to the Monte Carlo methods!

Now let's repeat the simulation many times and observe the distribution of estimates.

````python
n = 500
num_experiments = 1000

estimates = []

for _ in range(num_experiments):
    estimate = get_sample_expected_value(n)
    estimates.append(estimate)
````

This experiment shows how the Monte Carlo algorithm behaves across many **independent** simulations.

Each bar in the histogram represents a possible estimate of the average height obtained using random sampling. \

### 1.Discrete Probability Distributions:

A **discrete probability distribution** describes a random variable which can take on finite or countably infinite set of values. 

For discrete variables, probabilities are assigned directly to individual outcomes. \
The distribution is described by a **probability mass function (PMF)**:
$$
    p(x)=P(X = x)
$$

The **PMF** gives the probability that the random variable $X$ takes exactly the value $x$
There are some very important proparties of a **Probability Mass Function**:
A valid PMF must satisfy the following conditions:
1. Non-negativity
$$
    p(x)≥0
$$
- Probabilities cannot be negative.
2. Normalization
$$
    \sum_{x} p(x) = 1
$$
- The sum of the probabilities of all possible outcomes must equal 1 because one of the outcomes must occur.

These two properties guarantee that the function represents a valid probability distribution.

The **binomial distribution** is a discrete probability distribution that models the number of successes $x$ in a fixed number $n$ of independent trials (success/failure), where each trial has the same probability of success $p$.It is commonly used to calculate the probability of a specific number of outcomes.

The key and most important details about it are:

    Each trial has only two possibilities: success or failure (e.g., heads/tails, pass/fail).
- **Fixed Trials**:

- **Independent Trials**:

$$

$$

$$

bacause:
$$
    \binom{n}{k} = \frac{n!}{k!(n-k)!}(\text{binomial coefficient})
$$

- $n$ = number of trials

- $p$ = probability of success in a single trial

### 1.The Law of Large Numbers:
By definition, The Law of Large Numbers states that:
> As the number of independent samples increases, the sample average converges to the true expected value of the random variable.

If:
- $ X_1, X_2,..., X_n $ are independent and identically distributed (i.i.d.) random variables
- $ E(X) = \mu $

Then:
$$
    \frac{1}{N} \sum_{i=1}^{N}X_i \xrightarrow{N \to \infty} \mu
$$

where:
- $\frac{1}{N} \sum_{i=1}^{N}X_i$: The sample mean (average of $N$ observations).

- $\mu = E(X)$: The expected value (theoretical mean) of the population.

- $N$: Sample size

> As N increases, the average of the sample start to equall more and more to the real(expected) mean.

We can define even it more formally:
$$
    P\left( \lim_{n \to \infty} \overline{X}_n = \mu \right) = 1
$$

where:
- $\overline{X}_n = \frac{1}{N} \sum_{i=1}^{N}X_i$

Monte Carlo estimates:
$$
    E(X) \approx \frac{1}{N} \sum_{i=1}^{N}X_i
$$

#### The law of large numbers guarantees that this estimation will become more and more accurate as N gets larger!

Let's see it visually:

````python
mu = 5 # The expected, true mean
sigma = 2 # The sd

N = 100000

samples = np.random.normal(mu, sigma, N)

# the cumsum() function computes the cumulative sum of the elements in an array(basically it computes all of the elements before, plus the current and that way it makes the sum for each of them!)
# The process here is that it takes all of the values so far and computes their average, so that we can see it over the time and show it in the plot!
mean_over_time = np.cumsum(samples) / np.arange(1, N+1)

plt.figure(figsize=(10,6))

# Here we plot the mean over the time, to see how it is changing
plt.plot(mean_over_time, label="The Sample Mean")
plt.axhline(mu, linestyle='--', label="The True Mean")

plt.xlabel("Number of Samples")
plt.ylabel("Mean Value")

plt.title("The Law of Large Numbers")
plt.legend()

plt.show()
````

By, this we can assump that: \
As $N$ increases, the sample mean stabilizes more and more around the true mean!

We can also see how the mean vary over time:

````python
true_mean = 0
sd = 1

sizes_list = [10, 100, 1000, 10000, 100000]
mean_estimates = []

for size in sizes_list:
    samples = np.random.normal(true_mean, sd, size)
    estimated_mean = np.mean(samples)
    mean_estimates.append(estimated_mean)

df = pd.DataFrame({'n': sizes_list, 'estimate': mean_estimates})
print(df)
````

We can see that as the sizes of the simulations rise, the mean values become more and more equal to the true mean!

### 2.Central Limit Theorem (CLT):

By definition the **Central Limit Theorem** states that:
> if you take sufficiently large random samples from any population, the distribution of the sample means will approximate a **normal distribution**, **regardless** of the population's **original** distribution shape.

It is used to make inferences about populations, calculate probabilities, and construct confidence intervals.

>  As sample size increases, the sampling distribution of the mean becomes normal, centered at the population mean ($\mu$).

If:
$X_1, X_2,...,X_N$ be i.With mean $\mu$ and variance $\sigma$

Then:
$$
    \sqrt{N}(\bar{X} - \mu) \thicksim N(0, \sigma^2)
$$

or equivalently:
$$
    \bar{X} \thicksim N \left(\mu, \frac{\sigma^2}{N}\right)
$$

This explains why:
- why Monte Carlo estimates have a predictable error distribution
- why results become more stable with larger N
- why we can use normal approximations in statistics

So now lets actually see how from a **uniform distribution**, we will visualize a **normal** one:

````python
sample_size = 30
num_experiments = 5000

means = []

for _ in range(num_experiments):
    # Here we are sampling from a uniform flat distribution
    sample = np.random.uniform(0, 1, sample_size)
    estimated_mean = np.mean(sample)
    means.append(estimated_mean)

means = np.array(means)

plt.figure(figsize=(10,6))

plt.hist(means, bins=50, density=True)

plt.title("Central Limit Theorem: Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

plt.show()
````

#### We can see that the visualized distribution is a normal distribution!

**Small sample size = distribution not normal** \
**Larger sample size = approximately normal**

### Monte Carlo Error Analysis:

Monte Carlo methods rely on random sampling to approximate expectations.From understanding the error behavior behind the simulations, we can assess how **accurate and reliable** the estimates are.

Let:
$ X_1,X_2,...,X_N $ be i, samples of a random variable $X$ with mean $\mu = E[X]$ and variance $\sigma^2 = Var(X)$

The Monte Carlo estimation of the mean is:
$$
    \hat{\mu}_N = \frac{1}{N} \sum_{i=1}^{N}X_i
$$

The expected error is given by the standard deviation of the estimator:
$$
    Std(\hat{\mu}_N) = \frac{\sigma}{\sqrt{N}}
$$

- Error decreases with $\sqrt{N}$ = slow convergence
- Variance of the random variable directly affects the error
- Confidence intervals can be computed using CLT:
$$
    \hat{\mu}_N \pm z_{\alpha/2} \frac{\sigma}{\sqrt{N}}
$$

Where $z_{\alpha/2}$ comes from the standard normal distribution

For example lets simulate Monte Carlo estimation for $E[X]$ where $X \sim N(\mu=5, \sigma=2)$

````python
mu = 5
sigma = 2

# Here we get the sample sizes, from evenly spaced numbers on a log scale.
# The base values is by default 10, the first number will be: base ** start, and the last: base ** end
sample_sizes = np.logspace(1, 5, 50, dtype=int)
errors = []

# Making the simulations
for N in sample_sizes:
    samples = np.random.normal(mu, sigma, N)
    estimated_mean = np.mean(samples)
    
    # Here we calculate the absolute error
    error = abs(estimated_mean - mu)
    errors.append(error)

errors = np.array(errors)

plt.figure(figsize=(10,6))

# Here we plot the error vs the sample size:
plt.plot(sample_sizes, errors, label="Monte Carlo Error")
plt.plot(sample_sizes, sigma/np.sqrt(sample_sizes), '--', label=r'Theoretical $\sigma/\sqrt{N}$')

# In order to see the results clearly we plot the results on a log scale
plt.xscale('log')
plt.yscale('log')

plt.xlabel("Number of Samples (log scale)")
plt.ylabel("Absolute Error (log scale)")
plt.title("Monte Carlo Error Analysis")
plt.legend()

plt.show()
````

We can see that:
- Error decreases as $1/\sqrt{N}$
- Log-log scale shows a straight line -> power-law convergence
- Variance of $X$ controls the slope of convergence

> As the samples increases the errors decreases!

Now lets also observe the **confidence intervals** and how they apply for the even more accurate estimation of the expected value of a distribution:

By definition:
> A confidence interval (CI) gives a range of values that likely contains the true expected value.

As Gemini[9] says:
> "While a point estimate, such as the sample mean ($\bar{x}$), provides a single propolsal for the expected value, it rarely equals the true population parameter exactly. Confidence intervals improve this estimation by providing a range of plausible values that quantify the uncertainty around that point estimate."

From the CLT:
$$
    \hat{\mu} \sim \mathcal{N}\left(\mu, \frac{\sigma^2}{N}\right)
$$

Which states that:
- As sample size increases, the distribution of sample means approaches a normal distribution, allowing us to calculate CI bounds even if the underlying data is not normally distributed.

And also confirms that:
- The sample mean is a good estimator of the population mean and allows the formula $ SE = \frac{\sigma}{\sqrt{n}} $
which is very important for defining the precision of the interval.

As we have seen earlier the interval is defined as:
$$
    \hat{\mu}_N \pm z_{\alpha/2} \frac{\sigma}{\sqrt{N}}
$$

Where:
- $z_{\alpha / 2}$ is from the standard normal distribution
- $\sigma / \sqrt{N}$ standard error

The confidence levels are defined with the common values:
- 90% - 1.645
- 95% - 1.96
- 99% - 2.576

Now lets see some examples: \
Suppose $E[X], X \sim N(5,2)$

````python
mu = 5     # The mean
sigma = 2  # The standard deviation
N = 10000  # Sample size

# Sample from a normal distribution:
samples = np.random.normal(mu, sigma, N)


# Estimating the mean
sample_mean = np.mean(samples)

# Estimating the standard deviation (which is equal to sigma)
sample_std = np.std(samples)

# Standard error of the mean
standard_error = sample_std / np.sqrt(N)

z = 2.576 # for 99% confidence level

lower_bound = sample_mean - z * standard_error
upper_bound = sample_mean + z * standard_error

print("Estimate:", sample_mean)
print(f"95% Confidence Interval: {lower_bound}, {upper_bound}")
````

The interval contains the true mean with high probability \
Width depends on:
- variance
- number of samples

If this does not tell us nothing, lets visualize it:

````python
plt.figure(figsize=(10,6))

# Plot histogram of samples
plt.hist(samples, bins=50, density=True, alpha=0.6, label="Samples")

# Plot estimated mean
plt.axvline(sample_mean, linestyle='--', label="Estimated Mean")

# Plot confidence interval bounds
plt.axvline(lower_bound, linestyle=':', label="Lower CI")
plt.axvline(upper_bound, linestyle=':', label="Upper CI")

plt.title("Monte Carlo Estimate with 99% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Density")

plt.legend()
plt.show()
````

Now lets also see how, as the number of simulations increases, confidence interval becomes narrower

````python
sample_sizes = np.logspace(2, 5, 30, dtype=int)

intervals_widths = []

for N in sample_sizes:

    samples = np.random.normal(mu, sigma, N)

    sample_mean = np.mean(samples)
    sample_std = np.std(samples)

    standard_error = sample_std / np.sqrt(N)

    # CI width = 2 * margin of error
    ci_width = 2 * 1.96 * standard_error

    intervals_widths.append(ci_width)
    
````

````python
plt.figure(figsize=(10,6))

plt.plot(sample_sizes, intervals_widths)

plt.xscale("log")

plt.xlabel("Number of Samples")
plt.ylabel("Confidence Interval Width")

plt.title("Confidence Interval Width vs Sample Size")

plt.show()
````

We can see that the interval widths decreases as:
$$
    \frac{1}{\sqrt{N}}
$$

> More samples - more precise estimate!

So after these examples we can say that:
> Confidence Intervals tells how uncertain your Monte Carlo estimate is

Or more formally:
> Confidence intervals quantify the uncertainty of **Monte Carlo** estimates using the **Central Limit Theorem**

### Variance Reduction Techniques:

>The **Variance reduction** improves Monte Carlo efficiency **without** increasing the number of samples.It reduces the errors for the **same** number of samples, which is highly valuable in **computationally expensive simulations**.

For this technique we examine two different approaches: ***Antithetic Variates*** and ***Control Variates***

1.The *Antithetic Variates*:
- It uses negatively correlated samples to cancel variability
- If $X$ is a sample, also use $X' = f(1 - U)$ where $U \sim U[0,1]$

So let's see how it is working:

We estimate $E[e^X]$ where $X \sim U[0,1]$

So, without Variance Reduction:

````python
N = 10000
sample = np.random.uniform(0, 1, N)

# Apply the function f(x) = exp(x) to each random sample
# This is the random variable we want to estimate E[f(X)] for
Y = np.exp(sample)

estimate_standard_mean = np.mean(Y)

# Estimate the standard error of the mean
# Standard error = sample standard deviation / sqrt(number of samples)
# Measures uncertainty of the Monte Carlo estimate
std_standard_error = np.std(Y)/np.sqrt(N)

print("Standard Monte Carlo Estimate:", estimate_standard_mean)
print("Standard Error:", std_standard_error)
````

With Antithetic Variates:

````python
# Compute the antithetic samples
# Y1 is f(U), Y2 is f(1-U), which is negatively correlated with Y1
Y1 = np.exp(sample)
Y2 = np.exp(1-sample)

# Combine the two correlated estimates by averaging
# This reduces variance while keeping the estimator unbiased
Y_antithetic = (Y1 + Y2)/2

# Estimate the expected value using the mean of the antithetic samples
estimate_antithetic_mean = np.mean(Y_antithetic)

# Estimate the standard error of the antithetic estimator
# Should be smaller than standard Monte Carlo due to variance reduction
std_antithetic_error = np.std(Y_antithetic)/np.sqrt(N)

# Print the Monte Carlo estimate using antithetic variates and its standard error
print("Antithetic Variates Estimate:", estimate_antithetic_mean)
print("Antithetic Standard Error:", std_antithetic_error)
````

If we visualize it:

````python
plt.figure(figsize=(10,6))

plt.bar(['Standard MC','Antithetic'], [std_standard_error, std_antithetic_error], color=['blue','orange'])

plt.ylabel("Standard Error")
plt.title("Variance Reduction using Antithetic Variates")

plt.show()
````

We can see that:
- Standard error drops significantly
- Same number of samples = more precise estimate
- Antithetic variates exploit negative correlation to reduce variance

2.The *Control Variates*: \
The Control Variates method reduces variance by using an additional random variable which is **correlated** with the quantity we want to estimate and has a **known expected value**.

We introduce another variable Z such that:
- $E[Z]$ is known
- $Z$ is correlated with $X$

The direct Monte Carlo estimation is:
$$
    \hat{X} = \frac{1}{N} \sum X_i
$$

But by using the correlated variable we adjusted the estimator as:
$$
    \hat{X} = \frac{1}{N} \sum (X_i - c(Z_i - E[Z]))
$$

where:
- c is a constant chosen to minimize variance

The standart way of calculating c is:
$$
    c = \frac{Cov(X, Z)}{Var(Z)}
$$

Suppose we want $E[X^2]$ where $X \sim U[0,1]$ Let Let Z=X (with known mean  = 0.5) as **control variable**. \
In order to redece the variance the two variables X and Z should be really strongly connected. \
For this example they are, because $X^2$ and $X$ are strongly correlated!

Lets see it:

````python
N = 10000

# Generate random samples from Uniform[0,1]
# These represent the random variable X
X = np.random.uniform(0, 1, N)

# Here we define the target variable Y = X^2
# This is the quantity whose expected value we want to estimate
Y = X**2

# We define the control variable Z = X
Z = X

# Computing the standard Monte Carlo estimate of E[X^2]
estimate_standard_mean = np.mean(Y)

# Computing standard error of the estimate
std_standard_error = np.std(Y) / np.sqrt(N)


# Computing covariance between Y and Z
# The cov function, indicates the level to which two variables vary together
# If the resut of this is positive, it means that the two variables increase parallel
# if it is negative it means that as the one increases the other decreases
# If equal to zero, there is no linear relationship between them
cov_YZ = np.cov(Y, Z)[0, 1]

# Computing variance of Z
var_Z = np.var(Z)

# Computing optimal coefficient c
# This minimizes the variance of the estimator
c = cov_YZ / var_Z

# E[Z] = 0.5 for Uniform[0,1]
Y_control = Y - c * (Z - 0.5)

# Computing the improved estimate
estimate_control = np.mean(Y_control)

# Computing standard error after the variance reduction
std_control_error = np.std(Y_control) / np.sqrt(N)

print("Standard Monte Carlo Estimate:", estimate_standard_mean)
print("Standard Error:", std_standard_error)
print()
print("Control Variates Estimate:", estimate_control)
print("Control Variates Standard Error:", std_control_error)
````

We can see that:
- Both estimates are close to the true value: 0.333
- The Control Variates technique decreases the errors from the estimate

As we made a correlation between X and Z:
- Z explains part of the randomness in X
- add back the known expectation of Z

If the correlation is strong, it leads to large variace reduction and with this we can state that:
> Control variates remove noise using known information
