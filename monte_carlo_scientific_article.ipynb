## The Monte Carlo Method:

By definition:

> The Monte Carlo method is a computational algorithm that uses repeated random sampling to obtain numerical results, particularly for problems involving probability and uncertainty.It approximates solutions to complex integrals, optimizations, and predictions by simulating many possible outcomes based on probability distributions.

### What Is a Monte Carlo Simulation?
**A Monte Carlo simulation** is a way to model the probability of different outcomes in a process that cannot easily be predicted due to the intervention of **random variables**. It is a technique used to understand the impact of risk and uncertainty.

## Hisrory:

> __A game of solitaire, a case of insomnia, a chance meeting at a train station, and a penchant for gambling all contributed to the creation of one of the most influential computing tools in the world: the Monte Carlo method.__ 

#### Where does it get its name from?
The Monte Carlo simulation was named after the famous gambling destination in Monaco because chance and random outcomes are central to this modeling technique, as they are to games like roulette, dice, and slot machines.

#### Where and by who was the method invented?
The technique was initially developed by [Stanislaw Ulam](https://en.wikipedia.org/wiki/Stanis%C5%82aw_Ulam), a mathematician who worked on the [Manhattan Project](https://en.wikipedia.org/wiki/Manhattan_Project), the secret effort to create the first atomic weapon.He shared his idea with [John Von Neumann](https://en.wikipedia.org/wiki/John_von_Neumann), [Robert Richtmyer](https://en.wikipedia.org/wiki/Robert_D._Richtmyer) and [Nick Metropolis](https://en.wikipedia.org/wiki/Nicholas_Metropolis), colleagues at the Manhattan Project, and collaborated to refine the Monte Carlo simulation.

As Ulam recounts his inspiration:
> The first attempts I made to practice the Monte Carlo Method came from a question that occurred to me in 1946 while I was recovering from an illness and playing solitaire: what are the chances that a Canfield solitaire with 52 cards will come out successfully? After trying to estimate them through combinatorial calculations, I wondered whether a more practical method than “abstract thinking” might be to lay it out a hundred times and simply count the successful plays. \
With the emergence of **fast computers**, this approach seemed possible.Later in 1946, I described the idea to John von Neumann, and we began planning actual calculations.

#### What's the idea behind it?
The **core idea** behind the creation of the Monte Carlo method was to solve mathematically complex or intractable problems, particularly those involving **randomness or high-dimensional variables**.Invented during World War II, the goal was to harness the power of emerging early computers to transform deterministic problems into probabilistic ones (which can be estimated by **observing thousands of random outcomes**).


#### How does it work?
Unlike a **normal forecasting model**, Monte Carlo Simulation predicts a set of outcomes based on an estimated range of values versus a set of fixed input values. In other words, a Monte Carlo Simulation builds a model of possible results by leveraging a **probability distribution**, such as a **uniform or normal distribution**, for any variable that has inherent uncertainty. It, then, recalculates the results over and over, each time using a different set of random numbers between the minimum and maximum values. In a typical Monte Carlo experiment, this exercise can be repeated **thousands of times** to produce a large number of **likely outcomes**.

##### Basic Monte Carlo Simulation Algorithm:
The following pseudo-code algorithm demonstrates the general workflow of a Monte Carlo simulation. The goal is to estimate a quantity (such as an expected value, probability, or integral) by repeated random sampling.
```
    Input:
        N → number of simulations(A very big number)

    total ← 0

    for i = 1 to N do
        x ← generate random variable from probability distribution
        y ← evaluate function f(x)
        total ← total + y
    end for

    estimate ← total / N
    return estimate
```

Monte Carlo simulation approximates a value using:
1. **Random Sampling**
2. **Repeated Experiments**
3. **Averaging**
4. **Approximate Solution**


#### How to use Monte Carlo method?

Regardless of what tool you use or for what purpose you are using it, \
The Monte Carlo method involves **three basic steps**:

1. **Set up the predictive model** \
Identify both the dependent variable to be predicted and the independent variables (also known as the input, risk or predictor variables) that will drive the prediction.

2. **Specify probability distributions of the independent variables.** \
Use historical data and the analyst’s subjective judgment to define a range of likely values and assign probability weights for each.

3. **Run simulations repeatedly, generating random values of the independent variables.** \
Do this until enough results are gathered to make up a representative sample of the near infinite number of possible combinations.


After we know the workflow of the Monte Carlo method, lets try to give some example and represent that example with a basic algorithm which estimates specific results:

One simple example of a Monte Carlo Simulation is to consider calculating the probability of rolling two standard dice. There are 36 combinations of dice rolls. Based on this, we can manually compute the probability of a particular outcome.

Using a Monte Carlo Simulation, you can simulate rolling the dice 10,000 times (or more) to achieve more accurate predictions.

````python
import numpy as np
import matplotlib.pyplot as plt
````

By following the steps we have specified earlier, we will make an algorithm, which calculates the probability that the sum of the two dices would be 7. \
The theoretical probability of rolling 7 with two dice is:
$$ 6 / 36 \approx 0.1667 $$
But lets see it with code:

````python
# Number of simulations - should be a very big number in order the results to be more accurate!
n = 10000

# Counter for successful outcomes
success_count = 0

# Target event
target_sum = 7

for i in range(n):

    # Generate random values for independent variables
    die1 = np.random.randint(1,7)
    die2 = np.random.randint(1,7)

    # Evaluate the predictive model
    dice_sum = die1 + die2

    # Check if the event occurs
    if dice_sum == target_sum:
        success_count += 1

# Estimate probability
estimated_probability = success_count / n

print("Estimated probability:", estimated_probability)
````

#### Lets explain the algorithm a little bit more detailed:
##### 1.Number of Simulations (N):
` N = 10000 `

Larger values of N produce:

- more stable estimates
- smaller statistical error

##### 2.Random Sampling:
This function simulates rolling a die: `np.random.randint(1,7)`

- Each call generates an independent random value.

##### 3.Model Evaluation:
This step represents the system being studied: `dice_sum = die1 + die2`

##### 4.Event Detection:
This checks whether the desired outcome occurred: `if dice_sum == target_sum`

##### 5.Counting Successful Outcomes:
Each time the event occurs, the counter increases: `success_count += 1`

##### 6.Probability Estimation:
This computes the relative frequency of the event: `estimated_probability = success_count / N`


Now let's create the same algorithm but a little bit more compacted and using the build-in numpy functions for estimating average values:

````python
n = 10000

first_dice = np.random.randint(1, 7, n)
second_dice = np.random.randint(1, 7, n)

dice_sum = first_dice + second_dice

prob_of_sum_being_two = np.mean(dice_sum == 2)  # Calculates the arithmetic average of the specified elements.
print(prob_of_sum_being_two)
````

Or maybe being more then or equal to 10:

# The Mathematical foundations:

**The Monte Carlo Method** is built on several important ideas from probability theory and statistics.Although the method itself is simple in principle, using random sampling and repeated experiments to approximate a result, its reliability and accuracy depend on these underlying mathematical concepts.Understanding them, helps explain why Monte Carlo simulations work, how their results should be interpreted, and what factors influence their precision and accuracy.

In this section, we will introduce the **key mathematical ideas** which form the foundation of the Monte Carlo methods.

1. **Random Variables** \
    – variables whose values depend on random outcomes.
2. **Probability Distributions** \
    – the patterns which form the shape of randomeness.
3. **Expected Value** \
    – the average value of a random variable over many trials.
4. **Variance and Standard Deviation** \
    – measure how much simulation results vary.
5. **The Law of Large Numbers** \
    – the principle that the average of many random samples approaches the true expected value.
6. **The Central Limit Theorem** \
    – a result that explains how the distribution of sample averages behaves when many simulations are performed.
7. **Random Sampling** \
    – the process of generating values from probability distributions to represent uncertainty.
8. **Statistical Estimation** \
    – using simulation results to approximate unknown probabilities or quantities.

To start with, let's get acquainted with one of the many Monte Carlo techniques, which actually is a
foundational type of Monte Carlo method, and it is considered one of the most fundamental and frequently used techniques within the wide range of algorithms.

Take for an example, calculating the **average** height of the adult population of a country.It would require measuring the height of each person making up that population, summing up the numbers, and dividing them by the total number of people measured.This task might take a **very long time**!What we can do instead is take a sample of that population and compute its average height. It is unlikely to give you the exact average height of the entire population, however, this technique gives a result that is generally a good **approximation** of what that number might be.

Funnily enough, the **approximation** and the exact average of the entire population might sometimes be the same. This is only due to **chance**.In most cases, the numbers will be different.One question we might want to ask though, is **how different** - two key factors determine this difference, which we will address a little later?In fact, as the size of the sample **increases**, this **approximation** converges to the **exact** number(there is a specific law explaining this regularity and we will talk about it a little bit later).In other words, the error between the **approximation** and the actual result is getting smaller as the sample size increases!

Note that the height of a person is a random number.It can be anything really which is the nature of random things. Interestingly enough, a sum of random numbers is another random number. If you can't predict what the number making up the sum are, how can you predict the result of their sum? So, the result of the sum is also a random number - which random number points to the height of the population!

For a mathematician, the height of a population would be called a **random variable**, because the height among people making up this population varies randomly.We generally denote random variables with upper case letters. The letter $X$ is commonly used.

In statistics, the elements making up that population, which as suggested before are random, are denoted with a lower capital letter, generally $x$ as well.All these x's can also be seen as the possible outcomes of the random variable X. If we call X this random variable (the population height), we can express the concept of approximating the average adult population height from a sample with the following pseudo-mathematical formula:

$$
\text{Approximation}(\text{Average}(X)) = \frac{1}{N} \sum^{N}_{n = 1} x_n
$$

Which you can read as, the approximation of the **average value** of the random variable $X$, (the height of the adult population of a given country), is equal to the sum (the $\sum$ sign) of the height of $N$ adults randomly chosen from that population (the samples), divided by the number $N$ (the sample size). This, in essence, is what we call a **Monte Carlo approximation** - a method of **approximating a quantity of interest** — such as an **expected value**.

This **expected value** in statistics corresponds to the **average** of the random variable $X$ and it is called an expectation and is written as $E(X)$.

To summarize, Monte Carlo approximation (which is one of the MC methods) is a technique to approximate the expectation of random variables, using samples. It can be defined mathematically with the following formula:

$$
    E(X) \approx \frac{1}{N} \sum^{N}_{n = 1} x_n
$$

where:
- $X$ is the random variable, which points to the expected(average) height of the population($N$),
- $x_n$ are the random sampled heights,
- $N$ is the number of samples

The mathematical sign $\approx$ means that the formula on the right inside of this sign only gives an *approximation* of what the random variable $X$ expectation $E(X)$ is!

So now let's represent our example, with the population and their average height, with actual code and see what assumptions we can gather from it!:

First, as we have did earlier we should set up the predictive model and specify the random variable!

````python
# The average height of a person of the population.
mean_height = 170

# The standard deviation, which is a required to be specified, in order to sample from a normal distribution.
# We will talk about distributions later on this section!
sd_height = 10

# Random variable: H ~ Normal(mean, std)
# This function generates random values from a normal distribution.
def sample_height(n):  # Here n is the size of the sample.
    # We specify probability distributions of the independent variables.
    return np.random.normal(mean_height, sd_height, n)
````

Each generated value of the function is a height of a randomly selected person from the population!

````python
# We now approximate the expected value 𝐸[X] using random samples.
def get_sample_expected_value(N):
    heights = sample_height(N)
    expected_value = np.mean(heights)
    return expected_value
````

Now that we have defined the functions for creating the random samples and getting the expected value from them, we should make some experiments to see some results:

````python
# The size of the samples
n = 1000

estimate = get_sample_expected_value(n)

print("Estimated average height:", estimate)
print("True average height:", mean_height)
````

As we can see, the estimate(the expected value) is very close to the true mean.So here we got a good approximation of the real average height of the population!From this observation we can assump that when we sample **many** different random values(heights), and here the keyword is **many**, from a normal distribution, we can expect that the average(expected value) will be closely enough to the real mean of the sample(the population)!

Now let's make some experiment and see what will happen if, the size of the sample, differs:

````python
# Here we create an array with different sizes, which are going to be used for the creation of samples.
sample_sizes = np.arange(10, 5000, 10)  # The arrange function, accept for parameters: min/max values and a step by which the numbers to be taken!

estimates = []

# Here for each of the different sizes we calculate the estimated mean and append it to an array.
for n in sample_sizes:
    estimate = get_sample_expected_value(n)
    estimates.append(estimate)
````

So in order to see the difference in the expected values, based on the sizes applied for their samples, we will make a simple plot, to visualize the differences:

````python
# Adjusting the plot figure
plt.figure(figsize=(10,6))

# PLotting the estimated values of the samples and the corresponding sizes applied for them:
plt.plot(sample_sizes, estimates, label="Monte Carlo estimate")
# Drawing the line representing the true population mean - 170:
plt.axhline(mean_height, linestyle="--", label="True population mean")

plt.xlabel("Number of Samples")
plt.ylabel("Estimated Average Height")

plt.title("Monte Carlo Approximation of the Expected Height")
plt.legend()

plt.show()
````

This visualization shows how the **Monte Carlo approximation** approaches the true expected value as the number of samples increases.

With this experiment we can assump that:
> As the number of samples increases, the sample mean approaches the expected value.

This statement is strongly connected to something called: *The Law of large Numbers* - a super important mathematical concept, which we are going to explain a little bit later - this law is crucial to the Monte Carlo methods!

Now let's repeat the simulation many times and observe the distribution of estimates.

````python
n = 500
num_experiments = 1000

estimates = []

for _ in range(num_experiments):
    estimate = get_sample_expected_value(n)
    estimates.append(estimate)
````

````python
plt.figure(figsize=(10,6))

# This will make a histogram and this way we can easily see the frequency on witch, each of the expected means falls.
plt.hist(estimates, bins=30)
plt.axvline(mean_height, linestyle="--", label="True Mean")

plt.xlabel("Estimated Mean Height")
plt.ylabel("Frequency")

plt.title("Distribution of Monte Carlo Estimates")
plt.legend()

plt.show()
````

This experiment shows how the Monte Carlo algorithm behaves across many **independent** simulations.

Each bar in the histogram represents a possible estimate of the average height obtained using random sampling. \
With this said, we can say that:
- most estimtes cluster around the true value
- extreme errors are rare

Again this observations, leads to some very important concepts which we will discuss latter.
